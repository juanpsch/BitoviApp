```python
import chromadb

# 1. Conectar al cliente
client = chromadb.PersistentClient(path="../data/chroma_db") 
collection = client.get_collection(name="bitovi_full_docs")

# 2. Recuperar todos los documentos
results = collection.get(
    include=["metadatas", "documents"]
)

# 3. Deduplicar y Ordenar
seen_ids = set()
unique_sorted_docs = []

# Primero ordenamos por fecha (o date_ts para mayor precisión)
# zip nos permite mantener metadatos y documentos juntos
raw_docs = sorted(
    zip(results['metadatas'], results['documents']),
    key=lambda x: x[0].get('date', '0000-00-00'), 
    reverse=True
)

# Ahora filtramos los duplicados quedándonos solo con la primera aparición (la más nueva)
for metadata, document in raw_docs:
    # Usamos doc_id o la URL como identificador único
    doc_identifier = metadata.get('doc_id') or metadata.get('source')
    
    if doc_identifier not in seen_ids:
        unique_sorted_docs.append((metadata, document))
        seen_ids.add(doc_identifier)

# 4. Mostrar resultados para chequear
print(f"Total en DB: {len(results['documents'])} chunks")
print(f"Documentos únicos encontrados: {len(unique_sorted_docs)}\n")

for i, (meta, doc) in enumerate(unique_sorted_docs[:5]): # Mostramos los top 5 únicos
    print(f"{i+1}. [{meta.get('date')}] - {meta.get('title')[:80]}...")
    print(f"   ID: {meta.get('doc_id')} | URL: {meta.get('source')}\n")

```
